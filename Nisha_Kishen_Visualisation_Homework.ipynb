```python
# Calculate the mean of the price by cut
mean_price = diamonds_data.groupby('cut')['price'].mean()
```

```python
print("Mean price by cut:\n", mean_price)
```

```python
# Calculate the standard deviation of the price by cut
std_dev_price = diamonds_data.groupby('cut')['price'].std()
```

```python
print("\nStandard deviation of price by cut:\n", std_dev_price)
```

```python
#Calculate the median of the price by cut
median_price = diamonds_data.groupby('cut')['price'].median()
```

```python
print("Median price by cut:\n", median_price)
```

```python
# Calculate the mode of the price by cut (Note: Mode may return multiple values)
mode_price = diamonds_data.groupby('cut')['price'].agg(lambda x: x.mode().values[0])
```

```python
print("\nMode price by cut:\n", mode_price)
```

```python
# Calculate the variance of the price by cut
variance_price = diamonds_data.groupby('cut')['price'].var()
```

```python
print("\nVariance of price by cut:\n", variance_price)
```

```python
# Calculate the 25th and 75th percentiles of the price by cut
percentile_25 = diamonds_data.groupby('cut')['price'].quantile(0.25)
percentile_75 = diamonds_data.groupby('cut')['price'].quantile(0.75)
```

```python
print("\n25th percentile of price by cut:\n", percentile_25)
```

```python
print("\n75th percentile of price by cut:\n", percentile_75)
```

```python
# Calculate the range of the price by cut
price_range = diamonds_data.groupby('cut')['price'].max() - diamonds_data.groupby('cut')['price'].min()
```

```python
print("\nPrice range by cut:\n", price_range)
```

```python
# Calculate correlation between price and carat
correlation = diamonds_data['price'].corr(diamonds_data['carat'])
print("Correlation between price and carat:", correlation)
```

```python
# Calculate covariance between price and carat
covariance = diamonds_data['price'].cov(diamonds_data['carat'])
print("Covariance between price and carat:", covariance)
```

```python
# Function to calculate standard deviation
def calc_std_dev(data):
    mean = sum(data) / len(data)
    variance = sum((x - mean) ** 2 for x in data) / len(data)
    std_dev = variance ** 0.5
    return std_dev
```

```python
# Example usage
numbers = [1, 2, 3, 4, 5]
print("Standard deviation of numbers:", calc_std_dev(numbers))
```

```python
# 1. Boxplot of Diamond Price Vs Cut
# Create a boxplot of diamond price versus cut
sns.boxplot(x='cut', y='price', data= diamonds_data)# Add labels and title
plt.xlabel('Cut')
plt.ylabel('Price')
plt.title('Boxplot of Diamond Price Vs Cut')
# Show the plot
plt.show()
```

```python
# 2. Barplot: Average Diamond Price by Color
# Create a barplot of average diamond price by color
sns.barplot(x='color', y='price', data=diamonds_data, estimator='mean')
# Add labels and title
plt.xlabel('Color')
plt.ylabel('Average Price')
plt.title('Average Diamond Price by Color')
#show the plot
plt.show()
```

```python
# 3. Scatterplot with Facets: Scatterplot of Carat vs. Price by Cut"
# Create a scatterplot with facets of carat vs. price by cut
sns.lmplot(x='carat', y='price', data=diamonds_data, col='cut', col_wrap=3)
# Set labels and title
plt.xlabel('Carat')
plt.ylabel('Price')
plt.suptitle('Scatterplot of Carat vs. Price by Cut', y=1.02)
# Show the plot
plt.show()
```

```python
# 4. Pie Chart: Distribution of Diamond Cut
# Calculate the frequency of each diamond cut category
cut_distribution = diamonds_data['cut'].value_counts()
```

```python
# Create a pie chart to display the distribution of diamond cut categories
plt.figure(figsize=(8, 6))
plt.pie(cut_distribution, labels=cut_distribution.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Diamond Cut')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle
# Show the plot
plt.show()
```

```python
# Perform one-hot encoding for categorical columns
encoded_data = pd.get_dummies(diamonds_data, columns=['cut', 'color', 'clarity'])

# Select only numeric columns for correlation analysis
numeric_columns = encoded_data.select_dtypes(include=['int64', 'float64'])

# Create a correlation matrix
correlation_matrix = numeric_columns.corr()

# Create a correlation plot using Seaborn's heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Plot of Diamond Features')
plt.show()
```
